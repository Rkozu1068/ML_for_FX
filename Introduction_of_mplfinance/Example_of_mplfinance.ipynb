# How to use mplfinance on Jupyter Notebook

## Import library

```python
import mplfinance as mpf
import pandas as pd
```

## Data setup

```python
df_dairy = pd.read_csv('SP500_NOV2019_Hist.csv',index_col=0,parse_dates=True)
df_dairy.index.name = 'Date'
df_dairy.shape
```

```python
df_dairy.head(3)
```

```python
df_dairy.tail(3)
```

```python
mpf.plot(df_dairy)
```

```python
mpf.plot(df_dairy,type='candle')
```

```python
mpf.plot(df_dairy,type='line')
```

```python
mpf.plot(df_dairy,type='ohlc',mav=4)
```

```python
mpf.plot(df_dairy,type='candle',mav=(3,6,9))
```

```python
mpf.plot(df_dairy,type='candle',mav=(3,6,9),volume=True)
```

```python
mpf.plot(df_dairy,type='candle',mav=(3,6,9),volume=True,show_nontrading=True)
```

```python
df_intraday = pd.read_csv('SP500_NOV2019_IDay.csv',index_col=0,parse_dates=True)
df_intraday.head(3)
```

```python
df_intraday = df_intraday.drop('Volume',axis=1) # Volume is zero anyway for this intraday data set
df_intraday.index.name = 'Date'
df_intraday.shape
```

```python
df_intraday.head(3)
```

```python
df_intraday.tail(3)
```

```python
iday = df_intraday.loc['2019-11-06 15:00':'2019-11-06 16:00',:]
mpf.plot(iday,type='candle',mav=(7,12))
```

```python
iday2 = df_intraday.loc['2019-11-05':'2019-11-06',:]
mpf.plot(iday2,type='candle')
```

```python
mpf.plot(iday2,type='candle',show_nontrading=True)
```

```python
mpf.plot(df_intraday,type='ohlc',show_nontrading=True)
```

```python
mpf.plot(df_intraday,type='line') 
```
